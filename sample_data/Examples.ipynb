```python
%matplotlib inline
import matplotlib.pyplot as plt
from cloudy.cloudy import Cloudy
```

```python
# sample has 4 lines varying by temperature
sample = Cloudy('sample.grd', 'sample.ems', ['Temperature'])
```

```python
sample.df
```

```python
# labels is will be the same as the header of the df
sample.labels
```

```python
# Plot with labels
for label in sample.labels[:-2]:
    plt.plot(sample.grd[0], sample.df[label], label=label)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
```

```python
sample2 = Cloudy('sample_2.grd', 'sample_2.ems', ['hden', 'Temperature'])
```

```python
# Sample 2 has one line but at multiple hdens varying by temperature
sample2.df
```

```python
sample2.labels
```

```python
# Plot the line at different hdens
# Since the df differs by hden and temperature but mostly by temperature,
# it is better to display the data as two hdens varrying by temperature.
# sample2.grd[0] = sample.df['hden']
# This line deletes the duplicates and turns the set into a list for iteration
hdens = sorted(list(set(sample2.grd[0])))
for hden in hdens:
    df = sample2.df[sample2.df['hden'] == hden]
    data = df['H2   2.121m']
    temps = df['Temperature']
    label = "hden=%.1f" % (hden)
    plt.plot(temps, data, label=label)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
```
